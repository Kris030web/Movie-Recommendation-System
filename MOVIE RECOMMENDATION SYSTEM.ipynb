```python
# Convert release_date to datetime
movies_2['release_date'] = pd.to_datetime(movies_2['release_date'], errors='coerce')

# Create a release_year feature
movies_2['release_year'] = movies_2['release_date'].dt.year
```

```python
import ast

def parse_list(data):
    if isinstance(data, str):
        data = data.strip()
        # Handle empty strings
        if data == '' or data.lower() == 'nan':
            return []
        # Try to parse as a list of dicts
        try:
            parsed_data = ast.literal_eval(data)
            if isinstance(parsed_data, list):
                # Extract 'name' from dicts
                return [item['name'] for item in parsed_data if isinstance(item, dict) and 'name' in item]
            else:
                # Not a list, treat as a comma-separated string
                return [s.strip() for s in data.split(',')]
        except (ValueError, SyntaxError):
            # Not a list, try splitting on comma
            return [s.strip() for s in data.split(',')]
    elif isinstance(data, list):
        return data
    elif isinstance(data, (int, float)):
        # Convert numbers to strings and treat as a single-item list
        return [str(data)]
    else:
        # For any other type, convert to string and treat as a single-item list
        return [str(data)]


# Apply parsing to 'genres' and 'keywords'
movies_2['genres'] = movies_2['genres'].apply(parse_list)
movies_2['keywords'] = movies_2['keywords'].apply(parse_list)
```

>why we do parsing
The function converts stringified JSON-like lists into Python lists of genre or keyword names.
This step is crucial for content-based filtering, as it allows us to compare genres and keywords between movies.


```python
#Create a combined feature for content based filtering
# Function to clean and combine features into a single string
def create_soup(row):
    genres = ' '.join(row['genres']).lower().replace(' ', '')
    keywords = ' '.join(row['keywords']).lower().replace(' ', '')
    overview = row['overview'].lower()
    return f"{genres} {keywords} {overview}"

# Apply the function to create the 'soup' column
movies_2['soup'] = movies_2.apply(create_soup, axis=1)

# Verify the new 'soup' column
print(movies_2[['title', 'soup']].head())
```

The "soup" feature combines genres, keywords, and overview into a single text representation.
This helps in calculating movie similarity using text-based methods like TF-IDF and cosine similarity.


```python
#TEXT VECTORIZATION
#this converts the soup feature to numericsl data
from sklearn.feature_extraction.text import TfidfVectorizer

# Reduce to most popular 10K movies
movies_small = movies_2.sort_values('popularity', ascending=False).head(10000)

# Initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Fit and transform the 'soup' column to create the TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(movies_small['soup'])

# Print the shape of the TF-IDF matrix
print("Shape of TF-IDF Matrix:", tfidf_matrix.shape)
```

# explanation
TF-IDF (Term Frequency-Inverse Document Frequency) converts textual data into numerical vectors based on wordimportance.
This matrix will be used to compute movie similarity in content-based filtering.

```python
from sklearn.model_selection import train_test_split

# Select relevant columns for collaborative filtering
collab_features = ['id', 'vote_count']

# Split data into training and testing sets (80-20 split)
train_data, test_data = train_test_split(movies[collab_features], test_size=0.2, random_state=42)

# Verify the shapes of training and testing datasets
print("Training Data Shape:", train_data.shape)
print("Testing Data Shape:", test_data.shape)
```

>EXPLANATION
Collaborative filtering relies on numerical feedback (e.g., ratings). Here, we use vote_average as a proxy for user
ratings.
Splitting the data ensures that we can evaluate our model later.

```python
# Parameters
num_users = 1000
ratings_per_user = 10
min_vote_count = 0  # Start with 0, can adjust if needed

# Filter movies
movies_filtered = movies_2[movies_2['vote_count'] >= min_vote_count].copy()

# Check if there are enough movies after filtering
if len(movies_filtered) < ratings_per_user:
    print(f"Not enough movies with vote_count >= {min_vote_count}. Reducing min_vote_count.")
    # Lower threshold to median if not enough movies
    min_vote_count = int(movies_2['vote_count'].quantile(0.5))
    movies_filtered = movies_2[movies_2['vote_count'] >= min_vote_count].copy()
    if len(movies_filtered) < ratings_per_user:
        print(f"Still not enough movies. Only {len(movies_filtered)} movies available. Adjust ratings_per_user or review data.")
        # Adjust ratings_per_user to available movies
        ratings_per_user = len(movies_filtered)
        if ratings_per_user == 0:
            raise ValueError("No movies available to sample. Please check your data.")
        print(f"Setting ratings_per_user to {ratings_per_user}.")

# Initialize lists for synthetic ratings
user_ids = []
movie_ids = []
ratings = []

movie_index_list = movies_filtered.index.tolist()

# Create synthetic users and assign ratings
for user_id in range(1, num_users + 1):
    # If not enough movies to sample without replacement, sample with replacement
    replace_flag = len(movies_filtered) < ratings_per_user
    selected_indices = np.random.choice(movie_index_list, ratings_per_user, replace=replace_flag)
    user_ids.extend([user_id] * ratings_per_user)
    movie_ids.extend(movies_filtered.loc[selected_indices, 'id'].values)
    ratings.extend(np.random.randint(1, 6, ratings_per_user))  # Ratings from 1 to 5

# Create DataFrame
synthetic_ratings_df = pd.DataFrame({
    'userId': user_ids,
    'movieId': movie_ids,
    'rating': ratings
})

# Create User-Item Matrix
user_item_matrix = synthetic_ratings_df.pivot_table(
    index='userId',
    columns='movieId',
    values='rating'
).fillna(0)

print("User-Item Matrix Shape:", user_item_matrix.shape)
```

WEEK 4: MODEL BUILDING

CONTENT BASED FILTERING

```python
#we use cosine similarity on the matrix to get a similarity matrix
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
# 3. Recommendation Function (Content-Based)
def get_recommendations_content(title, cosine_sim=cosine_sim):
    idx = movies[movies['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]
```

```python
# Pick a movie ID that exists in the matrix
test_movie_id = user_item_matrix.columns[0]  # You can also pick any index here
test_movie_title = movies_filtered[movies_filtered['id'] == test_movie_id]['title'].values[0]

print(f"\nCollaborative Filtering Recommendations for '{test_movie_title}' (Movie ID: {test_movie_id}):")
recommendations = get_recommendations_collab(test_movie_id)

if recommendations:
    for i, movie in enumerate(recommendations, 1):
        print(f"{i}. {movie}")
else:
    print("No recommendations found.")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Compute item-item similarity matrix using cosine similarity
item_similarity = cosine_similarity(user_item_matrix.T)

# Transpose to get items (movies) as rows
item_similarity_matrix = cosine_similarity(user_item_matrix.T)


def get_recommendations_collab(movie_id, top_n=10):
    # Check if the movie ID exists in the matrix columns
    if movie_id not in user_item_matrix.columns:
        print(f"❌ Movie ID {movie_id} not found in the user-item matrix.")
        return []

    # Get index of the movie in the similarity matrix
    movie_idx = list(user_item_matrix.columns).index(movie_id)

    # Get similarity scores for that movie
    similarity_scores = item_similarity_matrix[movie_idx]

    # Get indices of top N similar movies (excluding itself)
    similar_movie_indices = np.argsort(similarity_scores)[::-1][1:top_n + 1]
    similar_movie_ids = [user_item_matrix.columns[i] for i in similar_movie_indices]

    # Map back to movie titles
    recommendations = movies_filtered[movies_filtered['id'].isin(similar_movie_ids)]['title'].tolist()

    return recommendations
```

```python
#CONTENT BASED EVALUATION
# Example: Content-Based Recommendations for a few movies
test_titles = ['The Dark Knight', 'Inception', 'Forrest Gump']

for title in test_titles:
    print(f"\nContent-Based Recommendations for '{title}':")
    try:
        recs = get_recommendations_content(title)
        for i, movie in enumerate(recs, 1):
            print(f"{i}. {movie}")
    except Exception as e:
        print(f"Error for '{title}': {e}")
```
